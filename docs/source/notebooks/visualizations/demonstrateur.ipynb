# Demonstrateur

```python
import json

import numpy as np
from matplotlib import pyplot as plt
import seaborn as sns
from IPython.display import HTML

from oze.utils.env_setup import load_config
from oze.inputs import load_building
from oze.buildings.capt import CAPT
from oze.visualization import HistAnimation
```

```python
# Switch to sns visualization and deactivate automatic plotting
sns.set()
plt.ioff()

# Load config
config = load_config(config_path="config.json")

# Load building
building = load_building(config, CAPT, params_path=config['path']['saving_directory'].joinpath('calib', 'optim.json'))

# Load history data
hist_data = np.load('hist_visualization_calib.npy')

# Select fewer visualizations
hist_data = hist_data[::5]
```

## Indoor temperature

```python
anim = HistAnimation(building, hist_data, label='t_int', display_score=True)
HTML(anim.animate().to_jshtml())
```

## Private consumptions

```python
anim = HistAnimation(building, hist_data, label='private', display_score=True)
HTML(anim.animate().to_jshtml())
```

## Cold consumption

```python
anim = HistAnimation(building, hist_data, label='cold', display_score=True)
HTML(anim.animate().to_jshtml())
```
